```python
def uniq_pattern_list(pat_list):
    uniq=[]
    for i in pat_list:
        if i not in uniq:
            uniq.append(i)
    return uniq
```

```python
upm=uniq_pattern_list(ppm)
len(upm)
```

```python
def uniq_pattern_frequency(pat_list,unq_pat):
    freq_pat=[]
    for i in unq_pat:
        count=0
        for j in pat_list:
            if i==j:
                count=count+1
        freq_pat.append([i,count])
    fc=[i[1]for i in freq_pat]
    fc=sorted(list(set(fc)),reverse=True)
    sfrp=[]
    for i in fc:
        for j in freq_pat:
            if i==j[1]:
                sfrp.append(j)
    return sfrp
```

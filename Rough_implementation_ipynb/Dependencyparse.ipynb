```python
doc=nlp(unicode(txt[0],"utf-8"))
a=[]
for token in doc:
    a.append([token.text,token.dep_, token.head.text, token.head.pos_,[child for child in token.children]])
#a={i[0]:i[1:] for i in a}
a
```

```python
fn=open("/home/judson/Desktop/sentenceSeg/DependencyParsed/ratio_decision_dep.txt","a")
for i in txt:
    doc=nlp(unicode(i,"utf-8"))
    st="{"
    a=[]
    for token in doc:
        a.append([token.text,token.dep_, token.head.text, token.head.pos_,[child for child in token.children]])
    for j in a:
        if type(j) is list:
            a1="["+str(",".join(str(k) for k in j))+"]"
            st=st+a1
        else:
            if str(j)=="" or str(j)==" ":
                st=st+" ,"
            else:
                st=st+str(j)+","
    st=st+"}"
    fn.writelines(str(st))
    
```

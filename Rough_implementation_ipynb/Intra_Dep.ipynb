```python
import spacy
nlp=spacy.load('en_core_web_sm')
def remove_stop_punct(s):
        from nltk.corpus import stopwords
        stopwords=stopwords.words('english')
        doc=nlp(s)
        #k=sorted([".","DT","TO","CC","IN"])
        st=" ".join(t.text for t in doc if (t.pos_ not in ["PUNCT"] )and(t.text.lower() not in stopwords))
        return st
#print remove_stop_punct(s)
```

```python
def dict_conv(st):
    doc=nlp(st)
    root=""
    for i in doc:
        if i.dep_=="ROOT":
            root= i.text
    d={tok.text:[tok.dep_,tok.head.text,tok.tag_,[child for child in tok.children]] for tok in doc}
    return [d,root]
```

```python
import networkx
def dfs(sd, start, end = "$$"):   
    h=networkx.Graph(sd)
    ptr=[]
    for pth in networkx.all_simple_paths(h,start,"$$"):
        ptr.append(pth[:-1])
    return ptr
```

```python
def parse_depend(dt,root):
    #md= {i:j[3] for i,j in dt.items()}
    #print dt
    sd=dict()
    sd["$$"]=[]
    for i,j in dt.items():
        if j[3]==[]:
            sd[str(i)]=["$$"]
        else:
            sd[str(i)]=[str(k) for k in j[3]]
    #print sd
    m=""
    #for i,j in dt.items():
    #    if j[0]=="ROOT":
    #        print j[0]
    #        m=str(i)
    #        break
    #print m
    #print sd[m]
    ml=dfs(sd,root,"$$")
    #print ml
    mpos=[[dt[str(j)][2] for j in i]for i in ml]
    mdep=[[dt[str(j)][0] for j in i]for i in ml]
    return mdep,mpos
```

```python
#input nested string list output string
def To_str(nlst):
    st="("
    for i in nlst:
        st=st+"("
        st=st+",".join(i)
        st=st+"),"
    return st[:-1]+")"
```

```python
f=open('/home/judson/Desktop/sentenceSeg/Dep_IntraSent/Dep_arguments',"a")
f1=open('/home/judson/Desktop/sentenceSeg/Labelled/arguments.txt',"r")
txt=f1.readlines()
for i in txt:
    s=remove_stop_punct(unicode(i,"utf-8"))
    #print s
    #t=nlp(s)
    dep=parse_depend(*dict_conv(s))
    st=To_str(dep)
    f.writelines(s+"\t"+st+"\t<ARGUMENTS>\n")
    #print (s+"\t"+st+"\t<ARGUMENTS>\n")
```

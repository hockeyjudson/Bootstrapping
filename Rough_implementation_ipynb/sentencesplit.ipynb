```python
def sentencespliter(sent):
    tr=sent.split(".")
    tr2=[]
    jump=0
    for  i in range(len(tr)):
        if jump==2:
            jump=1
            continue
        if len(tr[i].strip().split())==0:
            continue
        if i==len(tr)-1 and len(tr[i].split())<=6:
            tr2[-1]=tr2[-1]+" "+tr[i]
            continue
        if i==0:
            if len(tr[i].strip().split())>=6:
                #tr.append(tr[i])
                tr2.extend([tr[i]])
                continue
            else:
                #tr[i:i+1]=["".join(tr[i:i+1])]
                #tr2[i:i+1]=[" ".join(tr[i:i+2])]
                tr2.append(" ".join(tr[i:i+2]))
                jump=2
                continue
        if len(tr[i].strip().split())>=6:
            #tr.append(tr[i])
            tr2.append(tr[i])
        else:
            if jump==1 and (tr2[-1].split()[-1][0].isupper() or tr2[-1].split()[-2][0].isupper()):
                tr2[-1]=tr2[-1]+" "+tr[i]
                jump=0
            elif tr[i-1].split()[-1][0].isupper() or tr[i-1].split()[-2][0].isupper():
                #tr[i-1:i]=[" ".join(tr[i-1:i+1])]
                #tr2[i-1:i]=[" ".join(tr[i-1:i+1])]
                tr2.append(" ".join(tr[i-1:i+1]))
            else:
                #tr[i:i+1]=["".join(tr[i:i+1])]
                #tr2[i:i+1]=[" ".join(tr[i:i+2])]
                tr2.append(" ".join(tr[i:i+2]))
                jump=2
    return tr2
```

```python
def makesentencesplit(ts):
    tt=[]
    for j in ts:
        if len(j.strip().split())<=3:
            continue
        if(len(j.strip().split("."))<=1):
            if(j.strip().split()>=6):
                tt.append(j)
        elif(len(j.strip().split("."))>=2):
            k="".join([i+"." for i in j.split(".") if i!=""])
            tt.extend(sentencespliter(k.strip()))
    return tt
```

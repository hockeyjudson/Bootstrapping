```python
f=open('/home/judson/Documents/corpus/corpus/refined/06_1.txt')
txt=f.read()
txt.strip()
```

```python
from lexnlp.nlp.en.segments import sentences
sent_list=sentences.get_sentence_list(txt)
sent_list
#sl=[s for s in sent_list if 'find' in s.lower()]
#sl
```

```python
from lexnlp.extract.en.dates import get_raw_dates
dt=[s for s in sent_list if list(get_raw_dates(s))]
dt
```

```python
import spacy
nlp=spacy.load('en')
doc=nlp(dt[0])
for chunk in doc.noun_chunks:
    print(chunk)
```

```python
for s in doc.sents:
    print(s.root)
```

```python
for tok in doc:
    print(tok,tok.subtree)
```

```python
spacy.displacy.render(doc,style='dep',jupyter='True',options={'compact':True})
```

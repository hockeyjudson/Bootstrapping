```python
foldname="/home/judson/Desktop/sentenceSeg/Labelled1/"
f=open(foldname+"arguments.txt","r")
arg=f.readlines()
arg=[i.strip()+" <ARGUMENTS>\n" for i in arg[:100]]
arg
```

```python
f=open(foldname+"decision.txt","r")
dec=f.readlines()
dec=[i.strip()+" <DECISION>\n" for i in dec[:100]]
dec
```

```python
f=open(foldname+"facts.txt","r")
fac=f.readlines()
fac=[i.strip()+" <FACTS>\n" for i in fac[:100]]
fac
```

```python
f=open(foldname+"identify.txt","r")
iden=f.readlines()
iden=[i.strip()+" <IDENTIFY>\n" for i in iden[:100]]
iden
```

```python
f=open(foldname+"ratio.txt","r")
rat=f.readlines()
rat=[i.strip()+" <RATIO>\n" for i in rat[:100]]
rat
```

```python
import numpy as np
conc=arg+dec+iden+rat+fac
np.random.shuffle(conc)
conc
```

```python
conc_lab=[i.strip().split()[-1] for i in conc]
len(conc_lab)
conc_wr=[" ".join(i.strip().split()[:-1]+["\n"]) for i in conc]
len(conc_wr)
```

```python
#f=open("/home/judson/evaltext.txt","a")
f=open("/home/judson/evaltext3.txt","a")
for i in conc_wr:
    f.writelines(i)
f.close()
```

```python
f=open("/home/judson/prediction1800.csv","r")
ytest=f.readlines()
ytest=[i.strip().split(",")[-1].upper() for i in ytest]
len(ytest)
```

```python
from sklearn import  metrics
print(metrics.classification_report(conc_lab,ytest,target_names=["<ARGUMENTS>","<DECISION>","<FACTS>","<RATIO>","<IDENTIFY>"]))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
conf_mat = confusion_matrix(conc_lab, ytest)
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10,8))
sns.heatmap(conf_mat,cmap="Greens" ,annot=True, fmt='d',xticklabels=["<ARGUMENTS>","<DECISION>","<FACTS>","<RATIO>","<IDENTIFY>"], yticklabels=["<ARGUMENTS>","<DECISION>","<FACTS>","<RATIO>","<IDENTIFY>"])
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

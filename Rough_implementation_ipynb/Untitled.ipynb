```python
n=[]
import nltk
from nltk import word_tokenize, pos_tag, ne_chunk
def getnames(t):
    k=ne_chunk(pos_tag(word_tokenize(t)))
    for i in k:
        m=[]
        if type(i)==nltk.Tree:
            if i.label()=="PERSON":
                for j in i:
                    m.append(''.join(j[0]))
            n.append(m)
    return n
```

```python
h=[]
for i in sentences.get_sentence_list(txt):
    h.append(getnames(i))
print(h)
```

```python
import re
from os import walk
fs=[]
for (dp,dn,fn) in walk("/home/judson/Desktop/sentenceSeg/SentSegmented/"):
    fs.extend(fn)
print(fs)
```

```python
roles=["<arguments>","<facts>","<decision>","<ratio>"]
fn="/home/judson/Desktop/sentenceSeg/SentSegmented/"
arg=[]
dec=[]
fact=[]
rat=[]
for i in fs:
    f=open(fn+i,"r")
    txt=f.readlines()
    for j in txt:
        if len(j.strip().split())>3:
            if j.strip().split()[-1]== roles[0]:
                arg.append(j)
            elif j.strip().split()[-1]== roles[1]:
                fact.append(j)
            elif j.strip().split()[-1]==roles[2]:
                dec.append(j)
            elif j.strip().split()[-1]==roles[3]:
                rat.append(j)
            else:
                continue
```

```python
import pickle
pickle.dump(arg,open("arglist.pickle","wb"))
pickle.dump(dec,open("decisionlist.pickle","wb"))
pickle.dump(fact,open("factslist.pickle","wb"))
pickle.dump(rat,open("ratiolist.pickle","wb"))
```

```python
def frequent_patterns(ip_list):
    import numpy as np
    ls={}
    for i in ip_list:
        for j in i:
            j=",".join(j)
            if j not in ls:
                ls[j]=[j.split(","),1]
            else:
                ls[j][1]=ls[j][1]+1
    f=[j[1] for i,j in ls.items()]
    f_i=np.argsort(f)
    f_i=f_i[::-1]
    f_l=list(ls)
    freq=[ls[f_l[i]] for i in f_i]
    return freq
```

```python
arg_freq=frequent_patterns(arg)
arg_freq
```

```python
dec_freq=frequent_patterns(dec)
dec_freq
```

```python
rat_freq=frequent_patterns(rat)
rat_freq
```

```python
iden_freq=frequent_patterns(iden)
iden_freq
```

```python
fac_freq=frequent_patterns(fac)
fac_freq
```

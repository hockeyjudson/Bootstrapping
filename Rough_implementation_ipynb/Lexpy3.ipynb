```python
from lexnlp.extract.en.dates import get_raw_dates
def date_replace(txdate):
    if len(list(get_raw_dates(txdate)))==0:
        return txdate
    st=" date "
    a=list(get_raw_dates(txdate))
    ind=[]
    for j in a:
        ind.append(j[1])
    m=0
    stlen=len(st)
    for i in ind:
        txdate=txdate[:i[0]-m]+st+txdate[i[1]-m:]
        m=m+(i[1]-i[0]-stlen)
    return txdate
        
```

```python
k=[]
for ij in txt:
    k.append(date_replace(ij))
```

```python
from lexnlp.extract.en.regulations import get_regulations
a=[]
for i in k:
    if len(list(get_regulations(i)))==0:
        continue
    print(i,":",list(get_regulations(i)))
    a.append(list(get_regulations(i)))
    
```

```python
import stanfordcorenlp
import spacy
nlp=spacy.load('en_core_web_sm')
```

```python
s=u"Mr Robinson and Mr Lowth relied upon, the views formed by Mr Bradley and Mr Ong about whether the Agreement applied to work carried out on the X50 work site."
```

```python
def pos_dict(string):
    tag=["CC","CD","DT","EX","FW","IN","JJ","JJR","JJS","LS","MD","NN","NNS","NNP","NNPS","PDT","POS","PRP","PRP$","RB","RBR","RBS","RP","SYM",
    "TO","UH","VB","VBD","VBG","VBN","VBP","VBZ","WDT","WP","WP$","WRB"]
    pos=dict()
    for i in tag:
        pos[i]=0
    #print pos
    st=nlp(string)
    for i in st:
        if i.tag_ in pos:
            if pos[i.tag_]==1:
                continue
            else:
                pos[i.tag_]=1
    return sorted(pos.iteritems())
print pos_dict(s)
```

```python
a=[]
for  i in nlp(s):
    a.append(i.tag_)
    print i.text,i.tag_
print set(a)
```

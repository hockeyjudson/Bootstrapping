```python
f=open("/home/judson/Desktop/testing.txt","r")
txt=f.readlines()
txt=[i.strip() for i in txt if i.strip()!=("" or "." or '"')]
txt=[i for i in txt if i!='']
txt
```

```python
import re
ratio=[]
for i in txt:
    match_obs =re.search(".*provision.*", i, flags=re.IGNORECASE)
    if match_obs==None:
        continue
    ratio.append(match_obs.group(0))
len(ratio)
sumi=0
print ratio[9]
for i in ratio:
    print len(i.split())
    sumi+=len(i.split())
print ratio,(sumi/(len(ratio)-1))
```
